```python
import pandas as pd
df=pd.read_csv("/content/survey.csv")
```

```python
display(df.columns)
```

```python
df = df.drop('Country', axis=1)
display(df.head())
```

```python
df = df.drop('state', axis=1)
display(df.head())
```

```python
df = df.drop('mental_vs_physical', axis=1)
display(df.head())
```

```python
df = df.drop('coworkers', axis=1)
display(df.head())
```

```python
df = df.drop('no_employees', axis=1)
display(df.head())
```

```python
df = df.drop('anonymity', axis=1)
display(df.head())
```

```python
df = df.drop('leave', axis=1)
display(df.head())
```

```python
df = df.drop('supervisor', axis=1)
display(df.head())
```

```python
df = df.drop('comments', axis=1)
display(df.head())
```

```python
df = df.drop('obs_consequence', axis=1)
display(df.head())
```

```python
df.to_csv('data1.csv', index=False)
```

```python
def map_gender(gender):
    if gender == 'Male':
        return 1
    elif gender == 'Female':
        return 2
    else:
        return 0

df['Gender'] = df['Gender'].apply(map_gender)
display(df.head())
```

```python
display(df.isnull().sum())
```

```python
for col in ['self_employed', 'work_interfere']:
    mode_val = df[col].mode()[0]
    df[col].fillna(mode_val, inplace=True)

display(df.isnull().sum())
```

```python
df.to_csv('data2.csv', index=False)
```

```python
df = df[(df['Age'] >= 0) & (df['Age'] <= 100)]
display(df.head())
```

```python
df['self_employed'] = df['self_employed'].map({'Yes': 1, 'No': 0})
display(df.head())
```

```python
df['family_history'] = df['family_history'].map({'Yes': 1, 'No': 0})
display(df.head())
```

```python
df['treatment'] = df['treatment'].map({'Yes': 1, 'No': 0})
display(df.head())
```

```python
df['work_interfere'] = df['work_interfere'].map({'Often': 1, 'Never': 0, 'Rarely': 2, 'Sometimes': 3})
display(df.head())
```

```python
df['remote_work'] = df['remote_work'].map({'Yes': 1, 'No': 0})
display(df.head())
```

```python
df['tech_company'] = df['tech_company'].map({'Yes': 1, 'No': 0})
display(df.head())
```

```python
def map_benefits(benefit):
    if benefit == 'Yes':
        return 1
    else:
        return 0

df['benefits'] = df['benefits'].apply(map_benefits)
display(df.head())
```

```python
def map_care_options(care_option):
    if care_option == 'Yes':
        return 1
    else:
        return 0

df['care_options'] = df['care_options'].apply(map_care_options)
display(df.head())
```

```python
def map_wellness_program(program):
    if program == 'Yes':
        return 1
    else:
        return 0

df['wellness_program'] = df['wellness_program'].apply(map_wellness_program)
display(df.head())
```

```python
def map_seek_help(help_option):
    if help_option == 'Yes':
        return 1
    else:
        return 0

df['seek_help'] = df['seek_help'].apply(map_seek_help)
display(df.head())
```

```python
def map_mental_health_consequence(consequence):
    if consequence == 'Yes':
        return 1
    else:
        return 0

df['mental_health_consequence'] = df['mental_health_consequence'].apply(map_mental_health_consequence)
display(df.head())
```

```python
def map_phys_health_consequence(consequence):
    if consequence == 'Yes':
        return 1
    else:
        return 0

df['phys_health_consequence'] = df['phys_health_consequence'].apply(map_phys_health_consequence)
display(df.head())
```

```python
def map_mental_health_interview(interview):
    if interview == 'Yes':
        return 1
    else:
        return 0

df['mental_health_interview'] = df['mental_health_interview'].apply(map_mental_health_interview)
display(df.head())
```

```python
df = df.drop('phys_health_interview', axis=1)
display(df.head())
```

```python
df.to_csv('data3.csv', index=False)
```
